```python
import cv2
import numpy as np
import face_recognition
```

```python
# from google.colab.patches import cv2_imshow
```

```python
Shobhit = face_recognition.load_image_file("Students Images\Shobhit.jpg")
Shobhit = cv2.cvtColor(Shobhit, cv2.COLOR_BGR2RGB)

Shobhit_test = face_recognition.load_image_file("Students Images\Shobhit_test.jpg")
Shobhit_test = cv2.cvtColor(Shobhit_test, cv2.COLOR_BGR2RGB)



```

```python
faceLoc = face_recognition.face_locations(Shobhit)[0]
EncShobhit = face_recognition.face_encodings(Shobhit)[0]
cv2.rectangle(Shobhit, (faceLoc[3], faceLoc[0]), (faceLoc[1], faceLoc[2]),(255,0,255), 2)

faceLoc_test = face_recognition.face_locations(Shobhit_test)[0]
EncShobhit_test = face_recognition.face_encodings(Shobhit_test)[0]
cv2.rectangle(Shobhit_test, (faceLoc_test[3], faceLoc_test[0]), (faceLoc_test[1], faceLoc_test[2]),(255,0,255), 2)



print(faceLoc)
print(faceLoc_test)

```

```python
results = face_recognition.compare_faces([EncShobhit], EncShobhit_test, tolerance= 0.5 )
print(results)
```

```python
faceDis = face_recognition.face_distance([EncShobhit], EncShobhit_test )
print(faceDis)
```
